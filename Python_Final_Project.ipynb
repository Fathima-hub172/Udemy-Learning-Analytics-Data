```python
## encoding the special characters course title to the readable one

import pandas as pd
df = pd.read_csv(urls, encoding='utf-8')


mask = df['course_title'].str.contains(r'[^a-zA-Z0-9\s]', regex=True)
special_titles = df[mask]

print(special_titles[['course_title']].head(3000))
```

```python
## Handling missing values

# Checking missing values once again
print(df.isnull().sum())

df['num_reviews'] = df['num_reviews'].fillna(0)

df['level'] = df['level'].fillna("Unknown")
```

```python
## Handling Duplicates

df = df.drop_duplicates()
print(df.duplicated().sum())
```

```python
## Column Rename and dropping the  unwanted column if required

df.rename(columns={'num_subscribers':'subscribers','num_reviews':'reviews','num_lectures':'lectures'}, inplace=True)

```

```python
## Treating outliers

upper_limit = df['subscribers'].quantile(0.99)
df['subscribers'] = df['subscribers'].clip(upper=upper_limit)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.boxplot(x=df['subscribers'])
plt.title("Boxplot of Subscribers")
plt.show()
```

```python
## Checking skewness

print("Skewness (subscribers):", df['subscribers'].skew())
print("Skewness (reviews):", df['reviews'].skew())
print("Skewness (lectures):", df['lectures'].skew())


plt.figure(figsize=(12,5))

# Histogram of original subscribers
plt.subplot(1,2,1)
sns.histplot(df['subscribers'], bins=50, kde=True)
plt.title("Original Subscribers Distribution")

# Histogram of log-transformed subscribers
plt.subplot(1,2,2)
sns.histplot(df['subscribers_log'], bins=50, kde=True)
plt.title("Log-Transformed Subscribers Distribution")

plt.show()
```

```python
# Converting the data type

# Convert price to integer
df['price'] = df['price'].astype(int)

# Converting published_timestamp to datetime
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])

# Checking data types after conversion
print(df.dtypes)
```

```python
## date parts and derived field(Feature transformation)

# Extract year, month, day
df['year'] = df['published_timestamp'].dt.year
df['month'] = df['published_timestamp'].dt.month
df['day'] = df['published_timestamp'].dt.day

# Extract weekday (0=Monday, 6=Sunday)
df['weekday'] = df['published_timestamp'].dt.weekday



# Revenue estimate = price * subscribers
df['revenue_estimate'] = df['price'] * df['num_subscribers']

# Preview the new column
print(df[['course_title','price','num_subscribers','revenue_estimate']].head())



# Derived Parts

df['review_ratio'] = df['num_reviews'] / df['num_subscribers'].replace(0,1)
df['is_free'] = df['price'] == 0
df.info()
```
